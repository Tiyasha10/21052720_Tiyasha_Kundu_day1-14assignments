## Q1) What are Data Structures? List some of data structures and their use in real world application?

Data structures are fundamental components of computer science that organize and store data in a specific way to enable efficient operations and access. They provide a way to manage and manipulate data effectively, optimizing tasks such as insertion, deletion, and retrieval.

#### Here are some commonly used data structures along with their real-world applications:

__Arrays:__ Arrays store a fixed-size sequence of elements of the same data type. They are used in various applications, such as storing a list of student's grades, representing images, and implementing matrices.

__Linked Lists:__ Linked lists consist of nodes that contain data and a reference to the next node. They are useful for dynamic data structures where elements can be easily inserted or removed. Linked lists are employed in applications like implementing stacks, queues, and managing memory in operating systems.

__Stacks:__ Stacks follow the Last-In-First-Out (LIFO) principle. They are used for tasks like parsing expressions, implementing undo-redo functionality, and managing function calls in programming languages.

__Queues:__ Queues follow the First-In-First-Out (FIFO) principle. They are suitable for scenarios such as job scheduling, handling requests in web servers, and buffering data in networking.

__Trees:__ Trees are hierarchical data structures consisting of nodes connected by edges. They have applications in file systems, decision-making processes, organization structures, and searching algorithms like binary search trees.

## Q4) What is difference between ordered data type and unordered data type?

#### Ordered data types

- Elements have a specific order or sequence, and their position or index is meaningful. 
- Accessing elements can be done by their position or index.
- Examples of ordered data types in Python include lists and tuples.

#### Unordered data types

- Elements do not have a specific order, and their position or index is not meaningful. 
- Elements are typically accessed or retrieved based on their unique keys.
- Examples of unordered data types in Python include sets and dictionaries.
