```python
import numpy as np
import matplotlib.pyplot as plt
height = np.random.normal(140,20,1000)
plt.figure(dpi=200)
plt.title("Height distribution")
plt.xlabel("Height (cm) ")
plt.ylabel("Frequency")
plt.hist(height, bins=30,ec="gold")
```

The code has generated us a histogram plot of a random height distribution with a mean of 140 and a standard deviation of 20, using 1000 randomly generated height values.

## Q4) Write Down difference between interpreter and compiler

#### Intrepreter

- Translates program one statement at a time.

- Interpreters usually take less amount of time to analyze the source code.

- Programming languages like JavaScript, Python, Ruby use interpreters.

#### Compiler

- Scans the entire program and translates it as a whole into machine code.

- Compilers usually take a large amount of time to analyze the source code.

Programming languages like C, C++, Java use compilers.

## Q5) What is Data Structure, why we need one in programs?

A data structure is a way of organizing and storing data in a computer program. It defines how the data is organized, how it can be accessed, and what operations can be performed on it. Data structures provide a means to efficiently manage and manipulate data, enabling programmers to write more efficient and effective programs.
